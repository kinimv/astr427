# Question 1

```python
import math

from numpy import empty,arange
from matplotlib.pyplot import plot, xlabel, ylabel, show
```

```python
uH = 200
uL = 0
uM = 80
a = 0.2
L = 1
N = int(L/a)
h = 1e-4
epsilon = h / 1000

t1 = 0.01
t2 = 0.1
t3 = 0.4
t4 = 1.0
t5 = 10.0

tend = t5 - epsilon

T = empty(N+1, float)
T[0] = uH
T[N] = uH
T[1:N] = uL
Tp = empty(N+1, float)
Tp[0] = uH
Tp[N] = uH

t = 0.0
c = h / (a * a)

while t < tend:
    for i in range(1,N):
        Tp[i] = T[i] + c * (T[i+1] + T[i-1] - 2 * T[i])
    T,Tp = Tp, T
    t += h
    
    if abs(t - t1) < epsilon:
        plot(T)
        
xlabel("x")
ylabel("U")
show()
    
```

### Timestep 0.04

```python
uH = 200
uL = 0
uM = 80
a = 0.2
L = 1
N = int(L/a)
h = 1e-4
epsilon = h / 1000

ts = arange(0, 0.4, 0.04)

tend = 0.4 + 2*h

T = empty(N+1, float)
T[0] = uH
T[N] = uH
T[1:N] = uL
Tp = empty(N+1, float)
Tp[0] = uH
Tp[N] = uH

t = 0.0
c = h / (a * a)
stuff = empty(len(ts), float)
tt = 0

while t <= tend:
    for i in range(1,N):
        Tp[i] = T[i] + c * (T[i+1] + T[i-1] - 2 * T[i])
    T,Tp = Tp, T
    
    for t_s in ts:
        if (abs(t - t_s) < epsilon):
            stuff[tt] = T[2]
            tt += 1
            
    t += h
plot(ts, stuff)
xlabel("t")
ylabel("U")
show()
    
```
