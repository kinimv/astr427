```python
# Question 1
# EXERCISE 2.6
from math import pi, sqrt

M = 1.9891E30
G = 6.6738E-11
```

```python
l1 = float(input("Enter the closest distance from the sun: "))
v1 = float(input("Enter the linear velocity at the perihelion: "))
# Earth 1.4710E11 3.0287E4
# Haley 8.7830E10 5.4529E4
```

```python
#using the equation smaller root from the part above

a = 1
b = -2*G*M/(v1*l1)
c = -(v1**2 - (2*G*M/l1))

#linear velocity at aphelion
v2 = (-b - sqrt(b**2 - (4*a*c)))/2*a

#distance from the sun at aphelion
l2 = (l1*v1)/v2
```

```python
#semi major axis
a = 0.5*(l1 + l2)
#semi-minor axis
b = sqrt(l1*l2)
#orbital period
T = (2*pi*a*b)/(l1*v1)
#orbital eccentricity
e = (l2 - l1)/(l2 + l1)

print("Distance from the sun at aphelion is ", l2, "m")
print("Linear velocity at aphelion is ", v2, "m/s")
print("The orbital period is ", T, "seconds")
print("The eccentricity of orbit is ", e)
```

```python
import numpy as np
import matplotlib.pyplot as plt
 
# Reading from the .txt file
file = '/home/kinimv/code/astr427/stars.txt'

data = np.loadtxt(file,float)
temp = data[:,0]
appmag = data[:,1]
```

```python
# Plot
plt.figure(dpi=1500)
plt.scatter(temp, appmag, s=0.5)
plt.title('Hertzsprung–Russell diagram')
plt.xlabel('Temperature(in Kelvin)')
plt.xlim(15000,0)
plt.ylabel('Apparent Magnitude')
plt.ylim(20,-5)

#Rough location of main sequence
plt.plot([8000,5000],[1.5,5.2],'r-',lw=2, label='Main Sequence')
plt.plot([5000,3800],[5.2,12.5],'r-',lw=2,)

# Where M0 is mass of the sun
# Marking Stars with masses 0.75M0 
appmagfirst = appmag[appmag == 9.2]
tempfirst = temp[appmag == 9.2]
plt.scatter(tempfirst, appmagfirst, s=0.75, c='k', label='0.75 Mass of Sun') 

# Marking Stars with masses 1M0 
appmagfirst = appmag[appmag == 10.0]
tempfirst = temp[appmag == 10.0]
plt.scatter(tempfirst, appmagfirst, s=0.75, c='gold', label='1 Mass of Sun') 

# Marking Stars with masses 1.5M0 
appmagfirst = appmag[appmag == 11.22]
tempfirst = temp[appmag == 11.22]
plt.scatter(tempfirst, appmagfirst, s=0.75, c='orangered', label='1.5 Mass of Sun') 

plt.text(4000, 17, "low mass stars", fontsize=8)
plt.text(11000, 4, "high mass stars", fontsize=8)

plt.legend()
plt.show()
```
